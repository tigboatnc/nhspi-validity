# Investigating the Validity of the NHSPI Index as a predictor for COVID related statistics

`data science` `machine learning` `visualization`

> __NHSPI Index__ <br/>
The Index combines measures from multiple sources and perspectives to offer a broad view of the health protections in place for the nation as a whole and for each U.S. state. The Index identifies strengths as well as gaps in the protections needed to keep people safe and healthy in the face of large-scale public health threats, and it tracks how these protections vary across the United States and change over time.

## Case Peak Data

```python
case_peakDF = pd.read_csv('./data/case_peak_manual.csv',converters={"peak_dates": lambda x: x.strip("[]").replace("'","").split(", ")})
```

```python
case_peakDF
```

```python
# Removing states with less than 3 peaks 
case_peakDF['peak_times'] =  case_peakDF["peak_dates"].apply(lambda x: len(x))
case_peakDF = case_peakDF[case_peakDF['peak_times']>2]
case_peakDF = case_peakDF.reset_index(drop=True)
```

```python
# Sorting the peak array 1st 2nd third peek
def sort_peak_array(money_str):
    money_str.sort()
    return money_str

case_peakDF['peak_dates'] = case_peakDF['peak_dates'].apply(sort_peak_array)
```

```python
split_df = pd.DataFrame(case_peakDF['peak_dates'].tolist())
case_peakDF_processed = pd.concat([case_peakDF, split_df], axis=1)
```

```python
case_peakDF_processed = case_peakDF_processed.add_prefix('CP_')
```

```python
from datetime import datetime
```

```python
COVID_START_DATE = datetime.strptime("2019-12-12", "%Y-%m-%d")

case_peakDF_processed['CP_0'] = pd.to_datetime(case_peakDF_processed['CP_0'])
case_peakDF_processed['CP_0_diff'] = case_peakDF_processed['CP_0'] - COVID_START_DATE
case_peakDF_processed['CP_1'] = pd.to_datetime(case_peakDF_processed['CP_1'])
case_peakDF_processed['CP_1_diff'] = case_peakDF_processed['CP_1'] - COVID_START_DATE
case_peakDF_processed['CP_2'] = pd.to_datetime(case_peakDF_processed['CP_2'])
case_peakDF_processed['CP_2_diff'] = case_peakDF_processed['CP_2'] - COVID_START_DATE
```

```python
# Converting number of days to integer 
def convert_days_to_int(days_timestamp):
    return int(days_timestamp.days)

case_peakDF_processed['CP_0_diff'] = case_peakDF_processed['CP_0_diff'].apply(convert_days_to_int)
case_peakDF_processed['CP_1_diff'] = case_peakDF_processed['CP_1_diff'].apply(convert_days_to_int)
case_peakDF_processed['CP_2_diff'] = case_peakDF_processed['CP_2_diff'].apply(convert_days_to_int)
```

```python
case_peakDF_processed = case_peakDF_processed[['CP_state','CP_0_diff','CP_1_diff','CP_2_diff']]
```

```python


case_peakDF_processed.loc[:, case_peakDF_processed.columns!='CP_state'] = case_peakDF_processed.loc[:, case_peakDF_processed.columns!='CP_state'].apply(zscore)
```

```python
case_peakDF_processed
```

## NHSPI Index 

- Source [NHSPI Index](https://nhspi.org/tools-resources/)
- Xlsx Download Link : [NHSPI Index 2021 Resource List](https://nhspi.org/wp-content/uploads/2021/05/NHSPI_2021_Data_File_MMP_reviewed.xlsx) 

__Filtering__
- Only picked the domain sheet from the full spreadsheet for brevity. 
- Local : [nhspi index  - Domains.csv](./data/nhspi index  - Domains.csv)


__Quick Recap__
- __HSS__ - HEALTH SECURITY SURVEILLANCE
    - Actions to monitor and detect health threats, and to identify where hazards start and spread so that they can be contained rapidly.
- __CPE__ - COMMUNITY PLANNING & ENGAGEMENT COORDINATION
    - Actions to develop and maintain supportive relationships among government agencies, community organizations, and individual households; and to develop shared plans for responding to disasters and emergencies.
- __IIM__ - INCIDENT & INFORMATION MANAGEMENT
    - Actions to deploy people, supplies, money and information to the locations where they are most effective in protecting health and safety.
- __HD__ - HEALTH CARE DELIVERY
- __CM__ - COUNTERMEASURE MANAGEMENT
- __EOH__ - ENVIRONMENTAL & OCCUPATIONAL HEALTH


```python
nhspiDF = pd.read_csv('./data/nhspi index  - Domains.csv')
```

```python
nhspiDF.head()
```

# Experiment 1 

```python
nhspiDF.columns
```

```python
nhspiDF_process = nhspiDF[['STATE','YEAR','HSS','CPE','IIM','HD','CM','EOH']]
```

```python
stack = []
for year in nhspiDF_process['YEAR'].unique():
    df = nhspiDF_process[nhspiDF_process['YEAR'] == year]
    df = df.rename(columns={'HSS': f"HSS_{year}",
                          'CPE': f"CPE_{year}",
                          'IIM': f"IIM_{year}",
                          'HD': f"HD_{year}",
                         'CM': f"CM_{year}",
                         'EOH': f"EOH_{year}",})
    df = df.drop(columns=['YEAR'])
    stack.append(df)
    
```

```python
from functools import reduce
nhspiDF_process = reduce(lambda x, y: pd.merge(x, y, on = 'STATE'), stack)
```

```python
# Computing Zscores 
nhspiDF_process.loc[:, nhspiDF_process.columns!='STATE'] = nhspiDF_process.loc[:, nhspiDF_process.columns!='STATE'].apply(zscore)
```

```python
nhspiDF_process.head()
```

```python
caseDF_total_z = caseDF_total.apply(zscore)
```

```python
nhspiDF_process__caseDF_total_z =  nhspiDF_process.merge(caseDF_total_z, how='left',left_on='STATE',right_on='state')
```

```python
nhspiDF_process__caseDF_total_z.head()
```

```python
corrMatrix = nhspiDF_process__caseDF_total_z.corr()
corrMatrix_interest = corrMatrix[['new_case','new_death']]
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [20, 20]
sn.heatmap(corrMatrix_interest, annot=True)
plt.show()
```

```python
nhspiDF_process__caseDF_total_z.plot(x='STATE',y=['EOH_2020','new_death'])
```

```python
nhspiDF_process__caseDF_total_z.plot(x='STATE',y=['EOH_2019','new_death'],kind='bar')
```

# Experiment 3 : Census Clustering 

```python
from sklearn.cluster import KMeans
```

```python
censusDF = censusDF.drop(columns='stateName')
```

```python
censusDF.columns
```

```python
censusDF
```

```python
censusDF_process = censusDF.apply(lambda x: x.str.replace(',',''))
censusDF_process = censusDF_process.apply(lambda x: x.str.replace('%',''))
censusDF_process = censusDF_process.apply(lambda x: x.str.replace('$',''))
```

```python
censusDF_process.dropna(inplace=True)
```

```python
censusDF_process.columns
```

```python
censusDF_process = censusDF_process.astype({"Total Population":"float",
                         "Median Household Income":"float",
                        "Bachelor's Degree or Higher":"float",
                         "Employment Rate":"float",
                        "Total Housing Units":"float",
                         "Without Health Care Coverage":"float",
                        "Total Employer Establishments":"float",
                         "Total Households":"float",
                        "Hispanic or Latino (of any race)": "float"})
```

```python
t = list(censusDF_process.columns)
t.remove('stateCode')
```

### Cluster 3

```python
# Cluster 4 
c3 = []
censusDF_process_data = censusDF_process[t]
censusDF_process_stateIndex = censusDF_process['stateCode']

kmeans = KMeans(n_clusters=3, random_state=0).fit(censusDF_process_data)

censusDF_process_data['cluster'] = pd.Series(kmeans.labels_)
censusDF_process_data['state'] = censusDF_process_stateIndex

```

```python
c3.append(censusDF_process_data[censusDF_process_data['cluster']==0]['state'].to_list())
c3.append(censusDF_process_data[censusDF_process_data['cluster']==1]['state'].to_list())
c3.append(censusDF_process_data[censusDF_process_data['cluster']==2]['state'].to_list())
```

### Cluster 4 

```python
# Cluster 4 
c4 = []
censusDF_process_data = censusDF_process[t]
censusDF_process_stateIndex = censusDF_process['stateCode']

kmeans = KMeans(n_clusters=4, random_state=0).fit(censusDF_process_data)

censusDF_process_data['cluster'] = pd.Series(kmeans.labels_)
censusDF_process_data['state'] = censusDF_process_stateIndex
```

```python
c4.append(censusDF_process_data[censusDF_process_data['cluster']==0]['state'].to_list())
c4.append(censusDF_process_data[censusDF_process_data['cluster']==1]['state'].to_list())
c4.append(censusDF_process_data[censusDF_process_data['cluster']==2]['state'].to_list())
c4.append(censusDF_process_data[censusDF_process_data['cluster']==3]['state'].to_list())
```

### Cluster 7

```python
# Cluster 7 

c7 = [] 
censusDF_process_data = censusDF_process[t]
censusDF_process_stateIndex = censusDF_process['stateCode']

kmeans = KMeans(n_clusters=7, random_state=0).fit(censusDF_process_data)

censusDF_process_data['cluster'] = pd.Series(kmeans.labels_)
censusDF_process_data['state'] = censusDF_process_stateIndex

```
